```python
import pandas as pd

Tabla=pd.DataFrame({
'Nombres':['Rodrigo', 'Pedro', 'Ignacia', 'Luis', 'Valentino', 'Carlos', 'Martina', 'Javiera', 'Camilo', 'Bastián'],
'Edad':[21,23,20,24,22,20,22,21,20,20],
'Carrera':['Informática', 'Telecomunicaciones', 'Informática', 'Informática', 'Analista', 'Telecomunicaciones', 'Analista',
'Telecomunicaciones', 'Informática', 'Informática']
})
Tabla
```

3. ¿Cuántos estudiantes hay en cada carrera? Utiliza alguna función de Python para responder.

```python
Tabla['Carrera'].value_counts()
```

4. ¿Cuál es la edad más común entre los estudiantes?

```python
print(f'La edad mas comun entre los estudiantes es {Tabla["Edad"].mode()[0]} años')
```

5. Se espera que estos estudiantes egresen de su carrera en aproximadamente 3 años más. Agrega una nueva columna al DataFrame llamada "Edad al egresar" que contenga la edad que tendrá cada estudiante al finalizar sus estudios. Luego, muestra en la consola el DataFrame actualizado.

```python
Tabla['Edad al egresar']=Tabla['Edad']+3
print(Tabla)
```
